```python
# Import Libraries
import sqlite3
```

## Creating a Database

```python
# Connect SQL to a database file which will contain your databases. 
# Incase no such file exists, it'll create on its own
conn=sqlite3.connect('database.db')

# Get a cursor for it to execute commands
crsr=conn.cursor()

# Creata a Command
# This command will create a table named "USER" which will have the following attributes:
# 1. USERID - Can Support Character of 20 length
# 2. Password -Can Support Character of 15 length
# 3. Name - Can Support Chracter of 40 Length
# Primary Key - U_ID
crt_cmd = '''CREATE TABLE USER(
U_ID VARCHAR(20),
PWD VARCHAR(15) NOT NULL,
NAME VARCHAR(40) NOT NULL,
PRIMARY KEY (U_ID)
);'''

# Execute Command
crsr.execute(crt_cmd)

# Close Connection
conn.commit()
conn.close()
```

## Check if Database was Created

```python
conn=sqlite3.connect('database.db')
crsr = conn.cursor()

# Execute command to get all tables
crsr.execute("SELECT NAME FROM sqlite_master WHERE type='table'")

# Get the result of the command executed
result = crsr.fetchall()

# Print Results
print(result)

# Close Connection
conn.commit()
conn.close()
```

## Functions   

Define a function which will add a row into table "USER". This can be used during registration

```python
def add_user(userid,password,name):
    conn=sqlite3.connect('database.db')
    crsr = conn.cursor()
    
    crsr.execute("INSERT INTO USER VALUES(:USER_ID,:PASSWORD,:NAME)",
                {
                    "USER_ID":userid,
                    "PASSWORD":password,
                    "NAME":name
                })
    conn.commit()
    conn.close()
```

## FUNCTIONALITIES

1. ADDING USERS - Change the values of the variables and run the cell below

```python
userid = 5
password = "abcd"
name = "abcde"

add_user(userid,password,name)
```

## Result

```python
# Run this cell to see all users
conn=sqlite3.connect('database.db')
crsr = conn.cursor()

crsr.execute("SELECT * FROM USER")
# The result will be in the format - "USERID, PASSWORD, NAME"
result = crsr.fetchall()

print(result)
conn.commit()
conn.close()
```
